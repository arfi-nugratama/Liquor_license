```python
from lxml import html
from lxml import etree
import requests
import pandas as pd
```

```python
page = requests.get('https://azliquor.gov/query/results_activezip.cfm')
tree = html.fromstring(page.content)
tables = tree.xpath('.//table')
license_table = tables[1]
license_table
```

```python
# parse table
rows_parsed = []
for row_elements in license_table:
    row = []
    for col_elements in row_elements:
        cell_content = col_elements.text_content() 
        row.append(cell_content)
    rows_parsed.append(row)
```

```python
pd.DataFrame(rows_parsed[1:], columns=rows_parsed[0])
```
